```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

#Now we are going to use the sklearn library to load the dataset - iris.csv

```python
from sklearn.datasets import load_iris
data1 = load_iris()
data1
```

#Input Data:

```python
X = pd.DataFrame(data1['data'], columns = data1['feature_names'])
X
```

```python
#Output Columns
y = pd.DataFrame(data1['target'], columns = ['target_names'])
y
```

```python
print(X.shape)
print(y.shape)
```

Since the data values are not very far from each other hence we can skip the standardscaler process

#Train Test Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.75, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Loading the KNN ML Model
from sklearn.neighbors import KNeighborsClassifier
model1 = KNeighborsClassifier()   # Here the default value of k is 5. If you want different value, then specify in the parenthesis
```

```python
#Training the Model
model1.fit(X_train, y_train)
```

```python
#Predict
y_pred = model1.predict(X_test)
y_pred
```

```python
#Actual results
y_test
```

```python
# Model Evaluation
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
#Checking Log Loss
from sklearn.metrics import log_loss
```

```python
y_pred_proba = model1.predict_proba(X_test)
y_pred_proba
```

```python
print(log_loss(y_pred, y_pred_proba))
```

#Confusion Metrix

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
# Load the dataset into dataframe
df = pd.read_csv('/content/finance_dataset.csv')
df.head()
```

```python
# Custom mapping for classes
class_mapping = {'Low' : 0, 'Medium' : 1, 'High' : 2}
df['Financial_Status'] = df['Financial_Status'].map(class_mapping)
```

```python
df.head()
```

```python
df['Financial_Status'].value_counts()
```

Missing Values

```python
df.isnull().sum()
```

Duplicates

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
# Splitting the dataset into input and output
X = df.drop(columns = 'Financial_Status')
y = df[['Financial_Status']]
```

```python
print(X.shape)
X.head()
```

```python
print(y.shape)
y.head()
```

#Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler = StandardScaler()
X = Scaler.fit_transform(X)
```

```python
X = pd.DataFrame(X)
X
```

#Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#Model Training

```python
from sklearn.svm import SVC
```

```python
# Applying the Support Vector Machine classifier
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Predicting the results
y_pred = svm_classifier.predict(X_test)

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the SVM classifier: {accuracy * 100:.2f}%")
```

confusion_matrix

```python
confusion_matrix(y_test, y_pred)
```

classification_report

```python
print(classification_report(y_test, y_pred))
```

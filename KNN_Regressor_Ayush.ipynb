```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
dataset1 = pd.read_csv('/content/house_price_cwh.csv')
print(dataset1.shape)
dataset1.head()
```

```python
dataset1.isnull().sum()
```

```python
dataset1 = dataset1.dropna()
print(dataset1.shape)
dataset1.isnull().sum()
```

```python
dataset1.duplicated().sum()
```

#Splitting the data into input and output

```python
X = dataset1.drop(columns = 'MEDV')
print(X.shape)
X.head()
```

```python
y = dataset1[['MEDV']]
print(y.shape)
y.head()
```

#Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X = sc.fit_transform(X)
X = pd.DataFrame(data = X, columns = ['CRIM', 'ZN', 'INDUS', 'CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'])
print(X.shape)
X.head()
```

#Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.80, random_state = 42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#Model Training

```python
from sklearn.neighbors import KNeighborsRegressor
model1 = KNeighborsRegressor(n_neighbors = 5)
```

```python
#Train Model
model1.fit(X_train, y_train)
```

```python
#Predict Output
y_pred = model1.predict(X_test)
y_pred
```

```python
#Actual Price
y_test
```

#Performance Evaluation

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```

```python
#Our Model is 84 % accurate
```

#Loss

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
#Mean Squared Error:
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
#Root Meam Squared Error
import math
rmse = math.sqrt(mse)
rmse
```

```python
#Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)
mae
```
